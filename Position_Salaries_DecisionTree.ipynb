```python
#importing the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as myplot
```

```python
#importing the dataset
myData=pd.read_csv('Position_Salaries.csv')
myData
```

```python
x=myData.iloc[:,1:2].values
print(x)
y=myData.iloc[:,2].values
print(y)
```

```python
myplot.scatter(x,y,color='red')
myplot.xlabel('Position')
myplot.ylabel('salaries')
myplot.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
createModel=DecisionTreeRegressor(random_state=0)
createModel.fit(x,y)
```

```python
createModel.predict([[6.5]])
```

```python
import mpld3
%matplotlib notebook
x_grid=np.arange(min(x),max(x),0.001)
x_grid=x_grid.reshape((len(x_grid),1))
myplot.scatter(x,y,color='red')
myplot.plot(x_grid,createModel.predict(x_grid),color='blue')
myplot.xlabel('Position')
myplot.ylabel('salaries')
myplot.show()
```

```python
createModel.score(x,createModel.predict(x))
```
